### **Imports**

```python
# arrays:
import numpy as np       

# data manipulation:
import pandas as pd

# data visualization:
import matplotlib.pyplot as plt

# data visualization (source of some datasets):
import seaborn as sns

# high-quality exporting images
from IPython import display
display.set_matplotlib_formats('svg')
```

### **Configuration of matplotlib custom theme**

>**More information** can be found at: https://matplotlib.org/stable/tutorials/introductory/customizing.html.

```python
# customizations:
plt.rcParams['font.family'] = 'monospace'
plt.rcParams['font.size'] = 8
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.titlecolor'] = 'black'
plt.rcParams['axes.titlesize'] = 10
plt.rcParams['axes.spines.top'] = False 
plt.rcParams['axes.spines.right'] = False 
plt.rcParams['axes.labelcolor'] = 'darkgray'
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.edgecolor'] = 'darkgray'
plt.rcParams['axes.linewidth'] = 0.2
plt.rcParams['ytick.color'] = 'darkgray'
plt.rcParams['xtick.color'] = 'darkgray'
```

### **Loading the data**

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

### **Creating a new calculated column**

```python
tips['perc_bill'] = tips['tip'] / tips['total_bill'] * 100
tips.head()
```

```python
# obtaining the descriptive statistics to support the story to be told:
tips.describe()
```

### **Creating the chart**

```python
# creating a figure/axes:
fig, ax = plt.subplots()

# first graph (background):
plt.scatter(
      tips['total_bill'],
      tips['perc_bill'],
      color='darkgray',
      alpha=0.3     
  )

# second graph (highlighted information):
tips_highlight = tips.query('perc_bill >= 20')
plt.scatter(
      tips_highlight['total_bill'],
      tips_highlight['perc_bill'],
      color='#5D3FD3',
      alpha=0.7
  )

# Title and axis-labels:
plt.title('How many waiters receive tips that represent more\nthan 20% of total bill?', loc='left', pad=5)
plt.xlabel('Total bill ($)')
plt.ylabel('Tip: % of\ntotal bill ($)', rotation=0, labelpad=35)
# text elements:
plt.text(15, 35,
         'Only 1 of 4 waiters receive at least\n20% of  total bill as a tip',
         color = '#5D3FD3',
         verticalalignment='center',
         ha='left',
         size=7);

# Exporting it as a figure:
plt.tight_layout()
plt.savefig('Storytelling-Python-matplotlib.png', dpi=1200)
```

### **Other examples**

```python
# Creating a figure/axes:
fig, ax = plt.subplots()

# First chart:
plt.scatter(
      tips['total_bill'],
      tips['tip'],
      color='lightgray',
  )

# Second chart:
tips_hl = tips.query('tip >=8 | total_bill >= 40')  # hightlighted information
plt.scatter(
      tips_hl['total_bill'],
      tips_hl['tip'],
      color='Crimson',
      s=25
  )

# Linking the colorful points:
plt.plot(
      tips_hl['total_bill'],
      tips_hl['tip'],
      color='Crimson',
      alpha=0.7,
      ls='-.'
  )

# Title and axis labels:
plt.text(33, 9.75, 'These points are unusual', color='Crimson',  weight='bold')
plt.text(7.5, 7.25, 'These points are the ordinary ones', color='lightgray', weight='bold')
plt.title('Storytelling with data')
plt.xlabel('Total bill ($)')
plt.ylabel('Tip ($)')

# Exporting it as a figure:
plt.tight_layout()
plt.savefig('bonus_storytelling_e1.png', dpi=1200)
```
